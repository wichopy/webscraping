```python
monitor_prices = []
for container in containers:
    brand = container.div.div.a.img['title']
    product_name = container.findAll("a",{"class": "item-title"})[0].text
    #try:
    #print product_name
    price= container.findAll("li",{"class": "price-current"})[0].text.strip().split()[0]
    if price == "|":
        price= container.findAll("li",{"class": "price-current"})[0].text.strip().split()[1]
    #except:
    #    price = "not found"
    shipping= container.findAll("li",{"class": "price-ship"})[0].text.strip()
    monitor_prices.append([product_name,brand,price,shipping])
    
monitor_prices
```

```python
import unicodecsv as csv # to avoid encoding erors.


ofile  = open('mon_pricing.csv', "w+")
writer = csv.writer(ofile, delimiter='	', quotechar='"', quoting=csv.QUOTE_ALL)

for row in monitor_prices:
    writer.writerow(row)


ofile.close()


        
```
